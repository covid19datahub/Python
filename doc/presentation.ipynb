## Data acquiral

The main motivation to use COVID-19 Data Hub: `COVID19` (R) or `covid19dh` (Python) is the simplicity of data acquiral.

With a single call you can receive national statistics of COVID-19. The function lists the sources of the data on stdout.

````python
usa,usa_src = covid19("USA") # read US data
````

You receive a Pandas dataframe of data. 

````python
usa
````

````python
# plot confirmed
plt.plot(usa["date"], usa["confirmed"], "r")
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Confirmed cases of Covid-19 in USA", size=16)
plt.box(on=None)
plt.show()
````

````python
# plots
plt.plot(usa["date"], usa["deaths"], "k")
plt.plot(usa["date"], usa["recovered"], "y")
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Recovered and deaths of Covid-19 in USA", size=16)
plt.box(on=None)
plt.show()
````

Level sets the unit:

1. Country (default)
2. State | Region | Canton
3. City | Municipality | District

In the case of USA, level 1 stands for country (federation) and level 2 stands for states.

````python
usa2,usa2_src = covid19("USA", level = 2, verbose = False) # fetch US states data
````

The parameter `verbose` controls the message output.

````python
usa2
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)
# plot each state separately
for state, data in usa2.groupby("administrative_area_level_2"):
    data = usa2[usa2["administrative_area_level_2"] == state]
    ax.plot(data["date"], data["confirmed"], label = state)
    ax.text(data["date"].iloc[-1], data["confirmed"].iloc[-1], state, fontsize=7)
# format plot
fig.suptitle("Confirmed cases of Covid-19 in US States", size = 16)
#plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), fancybox = False, ncol = 8, prop={'size': 9})
plt.gcf().autofmt_xdate()
plt.box(on=None)
fig.show()
````

````python
# get maximal record
usa2[ usa2["confirmed"] == max(usa2["confirmed"]) ]
````

## Filtering

Except for country, the function also filters by start date and end date.

````python
x,src = covid19(["ES","italy"], start="2020-03-01", end="2020-04-30", verbose=False)
````

Previous query shows only data between `2020-03-01` to `2020-03-31` from Spain and Italy.

````python
x
````

````python
# plot each state separately
for country in x.groupby("administrative_area_level_1").groups:
    data = x[x["administrative_area_level_1"] == country]
    plt.plot(data["date"], data["confirmed"], label = country)
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Confirmed cases of Covid-19 in Italy and Spain", size = 16)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), fancybox = False, ncol = 8, prop={'size': 9})
plt.box(on=None)
plt.show()
````

## Vintage

The datahub generates daily snapshots of the datasets. Fetch the version from given day by setting `vintage` to `True` and `end` to the desired date.

````python
x, src = covid19("USA", end = "2020-04-22", vintage = True, verbose = False)
````

The vintage data are collected at the end of the day, but published with approximately 48 hour delay, once the day is completed in all the timezones.

Hence if `vintage = True`, but `end` is not set, warning is raised and `None` is returned.

````python
x, src = covid19("USA", vintage = True) # too early to get today's vintage
print(x, src)
````
